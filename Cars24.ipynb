```python
df.drop(['name'], axis=1, inplace=True)
```

```python
df = pd.get_dummies(df, columns=['fuel', 'seller_type'], dtype=int, drop_first=True)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df.transmission = encoder.fit_transform(df['transmission'])
```

```python
df.head()
```

```python
df.owner.unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder
order = ['Test Drive Car','Fourth & Above Owner','Third Owner','Second Owner','First Owner']
encoder = OrdinalEncoder(categories=[order])
df.owner = encoder.fit_transform(df[['owner']])
```

```python
df.head()
```

```python
df.mileage
```

```python
df.dropna(inplace=True)
```

```python
df.mileage = df.mileage.str.replace(" kmpl", "").str.replace(" km/kg", "").astype(float)
```

```python
df.head()
```

```python
df.engine = df.engine.str.replace(" CC", "").astype(int)
```

```python
df.head()
```

```python
df.max_power = df.max_power.str.replace(" bhp", "").astype(float)
```

```python
df.head()
```

```python
df.drop(['torque'], axis=1, inplace=True)
```

```python
df.head()
```

```python
x = df.drop(['selling_price'], axis=1)
y = df.selling_price
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, y_pred)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
