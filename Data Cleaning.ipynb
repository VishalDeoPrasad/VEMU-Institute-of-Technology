```python
import pandas as pd
import numpy as np

data = {
    "Name" : ['Rahul', 'Priya', 'Riya', '-'],
    "Gender" : ["Male", "F", "female", "Male"],
    "Age" : [18, 19, 18, 20],
    "Marks": [88, np.nan, 76, 54],
    "City" : ['Bangalore', 'Mumbai', 'Delhi', 'Banglore'],
    "Attendence": [92, 85, np.nan, 90]
}
df = pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

```python
val = df.Marks.median()
```

```python
df.Marks.fillna(val, inplace=True)
```

```python
df
```

```python
val = df.Attendence.mean()
df.Attendence.fillna(val, inplace=True)
```

```python
df
```

```python
df.Gender = df.Gender.str.lower().replace({'f':'female', 'm':'male'})
```
